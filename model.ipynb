```python
def bool_to_int(boolean):
    if boolean == True:
        boolean = 1
        return boolean
    elif boolean == False:
        boolean = 0
        return boolean
```

```python
df18['is_new'] = df18['is_new'].apply(bool_to_int)
df18['garden'] = df18['garden'].apply(bool_to_int)
df18['terrace'] = df18['terrace'].apply(bool_to_int)
df18.head()
```
